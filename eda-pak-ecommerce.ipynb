# <b>EDA on Pakistan's Largest E-Commerce Dataset

### <b> About Dataset </b>
This is the largest retail e-commerce orders dataset from Pakistan. It contains half a million transaction records from March 2016 to August 2018. The data was collected from various e-commerce merchants as part of a research study.

### <b> Content </b>
<b>Geography:</b> Pakistan

<b>Time period:</b> 03/2016 – 08/2018

<b>Unit of analysis:</b> E-Commerce Orders

<b>Dataset:</b> The dataset contains detailed information of half a million e-commerce orders in Pakistan from March 2016 to August 2018. It contains item details, shipping method, payment method like credit card, Easy-Paisa, Jazz-Cash, cash-on-delivery, product categories like fashion, mobile, electronics, appliance etc., date of order, SKU, price, quantity, total and customer ID. This is the most detailed dataset about e-commerce in Pakistan that you can find in the Public domain.

<b>Variables:</b> The dataset contains
- Item ID
- Order Status (Completed, Cancelled, Refund)
- Date of Order
- SKU
- Price
- Quantity
- Grand Total
- Category
- Payment Method
- Customer ID.

# Functions made for plot customizations and style

```python
# Function for setting style and size of plot
def set_size_style(width, height, style):
    plt.figure(figsize=(width, height))
    sns.set_style(style)

# Function for customizating the plot
def customize_plot(plot, title:str, xlabel:str,  ylabel:str, title_font:int, label_font:int):
    plot.set_title(title, fontsize = title_font, weight='bold')
    plot.set_xlabel(xlabel, fontsize = label_font, weight='bold')
    plot.set_ylabel(ylabel, fontsize = label_font, weight='bold')
    
```

# What is the best-selling product categories among Pakistani online shoppers?

```python
pak_ecommerce_data['category_name'].value_counts()
```

```python
# Replacing all '\N' values with 'Others' in category_name
pak_ecommerce_data['category_name'].replace('\\N','Others', inplace=True)
```

```python
# Making a count chart by counting all orders by category
set_size_style(12,6,'darkgrid')

ax = sns.countplot(data=pak_ecommerce_data, x='category_name', hue='category_name', palette='husl')

customize_plot(ax, 'All selling categories among Pakistani shoppers', 'Category', 'Items sold', 12, 10)
for container in ax.containers:
    ax.bar_label(container)
ax.tick_params('x', rotation=90)
```

- Mobiles and Tablets is the best selling category, along with Men and Women's Fashion and appliances.

# Sale trends of previous years, in which months there are high sales?

```python
# Histogram of order frequency by date
set_size_style(10,5,'whitegrid')
ax = sns.histplot(data=pak_ecommerce_data, x='created_at', bins=20, color='orange', kde=True)
customize_plot(ax, 'Distribution of orders by date', 'Date', 'Order frequency', 12,10)
```

```python
# Grouping dataset by order date and summing the grand total
total_sales_each_day = pak_ecommerce_data.groupby('created_at')[['grand_total']].sum().reset_index()
```

```python
# Plotting a line plot with total sales on y-axis and date on x-axis
set_size_style(16,5,'darkgrid')
ax = sns.lineplot(data=total_sales_each_day, x='created_at', y='grand_total' ,errorbar=None)
ax.set_yticks(range(0,160000000+1,20000000))
ax.set_yticklabels(['0M', '20M', '40M', '60M', '80M', '100M', '120M', '140M', '160M'])
customize_plot(ax, 'Sales trend (All time)', 'Date', 'Total sales',13,10)

plt.show()
```

- There are hikes especially at the end of years. Hike indicates a lot of sale, these are due to promotional and sales campaigns of different E-Commerce companies. Let's have a closer look at each year.

```python
# Extractind date and month component from order date
total_sales_each_day['year'] = total_sales_each_day['created_at'].dt.year
total_sales_each_day['month'] = total_sales_each_day['created_at'].dt.month

# Subsetting the data in each different year
sale_trend_16 = total_sales_each_day[total_sales_each_day['year'] == 2016]
sale_trend_17 = total_sales_each_day[total_sales_each_day['year'] == 2017]
sale_trend_18 = total_sales_each_day[total_sales_each_day['year'] == 2018]
```

```python
# Line plot for 2016
set_size_style(16,5,'darkgrid')
ax = sns.lineplot(data=sale_trend_16, x='created_at', y='grand_total', color='purple', markers=True)
ax.set_yticks(range(0,70000000+1,10000000))
ax.set_yticklabels(['0M', '10M', '20M', '30M', '40M', '50M', '60M', '70M'])
customize_plot(ax, 'Sales trend (2016)', 'Date', 'Total sales',13,10)

plt.show()
```

```python
# Line plot for 2017
set_size_style(16,5,'darkgrid')
ax = sns.lineplot(data=sale_trend_17, x='created_at', y='grand_total', color='teal', markers=True)
ax.set_yticks(range(0,160000000+1,20000000))
ax.set_yticklabels(['0M', '20M', '40M', '60M', '80M', '100M', '120M', '140M', '160M'])
customize_plot(ax, 'Sales trend (2017)', 'Date', 'Total sales',13,10)

plt.show()
```

# How does the choice of payment method (credit card, Easy-Paisa, Jazz-Cash,cash-on-delivery) influence the completion and satisfaction of e-commerce orders?

```python
pak_ecommerce_data['status'].value_counts()
```

```python
# Making four main categories in order status out of all other categories i.e complete, refund, canceled, others
pak_ecommerce_data['status'].replace(['received', 'cod', 'paid'],'complete', inplace=True)

pak_ecommerce_data['status'].replace('order_refunded', 'refund', inplace=True)

pak_ecommerce_data['status'].replace('closed', 'canceled', inplace=True)

pak_ecommerce_data['status'].replace(['payment_review', 'pending', 'processing', 'holded', 'fraud',
                                      'pending_paypal', 'exchange', '\\N'], 'others', inplace=True)
```

```python
pak_ecommerce_data['status'].value_counts()
```

```python
pak_ecommerce_data['payment_method'].value_counts()
```

```python
# Making four main categories in payment methods out of all other categories.
pak_ecommerce_data['payment_method'].replace(['Payaxis', 'internetbanking','mcblite', 'ublcreditcard', 
                                              'bankalfalah'], 'Banks/Cards/Wallets', inplace=True)

pak_ecommerce_data['payment_method'].replace(['easypay_voucher', 'Easypay', 'Easypay_MA'], 'Easypaisa', inplace=True)
pak_ecommerce_data['payment_method'].replace(['jazzwallet', 'jazzvoucher'], 'JazzCash', inplace=True)

pak_ecommerce_data['payment_method'].replace(['cashatdoorstep', 'cod'], 'Cash on Delivery', inplace=True)

pak_ecommerce_data['payment_method'].replace(['customercredit', 'apg', 'mygateway', 'productcredit',
                                              'marketingexpense', 'financesettlement'], 'Others', inplace=True)
```

```python
pak_ecommerce_data['payment_method'].value_counts()
```

```python
# Visualizing count of payment methods
set_size_style(7,5,'whitegrid')
ax = sns.countplot(data=pak_ecommerce_data, x='payment_method', edgecolor = 'k')
customize_plot(ax,'Payment methods count', 'Method', 'Count', 12,10)
```

```python
# Visualizing count of order status
ax = sns.countplot(data=pak_ecommerce_data, x='status', edgecolor='k')
customize_plot(ax,'Orders status', 'Status', 'Count', 12,10)
```

```python
# There is a very small value of 'others' in status so removing it won't make any difference
set_size_style(10,6,'whitegrid')
ax = sns.countplot(data=pak_ecommerce_data[pak_ecommerce_data['status'] != 'others'], x='payment_method', hue='status', palette='tab10')

customize_plot(ax,'Payment methods with their order statuses', 'Payment methods','Order count',12,10)
```

- There is a high cancellation rate in online payments.
- Easypaisa and Banks/Cards/Wallets has the highest rate of cancellation.
- Cash on Delivery orders are the most successful with most them marked as completed.

```python
# Checking cross table of payment method and status
pd.crosstab(pak_ecommerce_data['payment_method'], pak_ecommerce_data['status'])
```

```python
# Subsetting canceled orders from all orders
canceled_orders = pak_ecommerce_data[(pak_ecommerce_data['status'] == 'canceled') &
                                     (pak_ecommerce_data['payment_method'] != 'Others')]

# Subsetting completed orders from all orders
completed_orders = pak_ecommerce_data[(pak_ecommerce_data['status'] == 'complete') &
                                      (pak_ecommerce_data['payment_method'] != 'Others')]

complete_order_methods = pd.crosstab(completed_orders['created_at'], completed_orders['payment_method'])
canceled_order_methods = pd.crosstab(canceled_orders['created_at'], canceled_orders['payment_method'])
```

```python
ax = complete_order_methods['2016' : '2016'].plot(figsize=(16,5))
customize_plot(ax, 'Completed orders in 2016', 'Months', 'Orders', 13,10)
```

```python
ax = canceled_order_methods['2016' : '2016'].plot(figsize=(16,5))
customize_plot(ax, 'Canceled orders in 2016', 'Months', 'Orders', 13,10)
```

```python
ax = complete_order_methods['2017' : '2017'].plot(figsize=(16,5))
customize_plot(ax, 'Completed orders in 2017', 'Months', 'Orders', 13,10)
```

```python
ax = canceled_order_methods['2017' : '2017'].plot(figsize=(16,5))
customize_plot(ax, 'Canceled orders in 2017', 'Months', 'Orders', 13,10)
```

```python
ax = complete_order_methods['2018' : '2018'].plot(figsize=(16,5))
customize_plot(ax, 'Completed orders in 2018', 'Months', 'Orders', 13,10)
```

```python
ax = canceled_order_methods['2018' : '2018'].plot(figsize=(16,5))
customize_plot(ax, 'Canceled orders in 2018', 'Months', 'Orders', 13,10)
```

- Customers find it more convenient to pay cash on delivery for the order, so this method is quite stable and also have very less order cancellations.
- There are more cancellations in the online payment methods.
- People prefer Cash on delivery more, but you can see hikes in online payments, it may be due to discount vouchers, sales and different campaigns which are eligible for mobile wallets and internet banking.
